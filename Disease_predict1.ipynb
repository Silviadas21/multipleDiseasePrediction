```python
traindata = traindata.drop("Unnamed: 133", axis=1)
```

```python
Y = traindata[["prognosis"]]
X = traindata.drop(["prognosis"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
rfc = RandomForestClassifier(random_state=42,max_depth=6, min_samples_split=18,n_estimators=100)

#best_rfc = grid_search.best_estimator_
#rfc.fit(xtrain,ytrain.values.ravel())
model_rfc = rfc.fit(xtrain,ytrain.values.ravel())
tr_pred_rfc = model_rfc.predict(xtrain)
ts_pred_rfc = model_rfc.predict(xtest)

print("training accuracy is:",accuracy_score(ytrain,tr_pred_rfc))
print("testing accuracy is:",accuracy_score(ytest,ts_pred_rfc))
```
